```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Requests it's allow you to send Http request using python.
# BeautifulSoup Library is use to pull data out of Html and Xml Files. 
# Importing pandas Library to allows you to access many of NumPy methods with less no of code.
```

```python
link='https://www.ecb.europa.eu/stats/policy_and_exchange_rates/euro_reference_exchange_rates/html/index.en.html'

# From this url we are going to scrape the data
```

```python
response=requests.get(link)
response

# If you get a response 200 then only you can scrapped the data from this website.
```

```python
soup=BeautifulSoup(response.text)
soup

# Beautifulsoup is taking that text file to get the data in Html formate.
```

```python
table=soup.find('table', class_ = 'forextable')
table

# Taking common attribute from the Html file and finding the table and that table class.
```

```python
table1=soup.find('tbody')
table1

# In the table we are finding the table body tag (tbody).
```

```python
rows=table1.find_all('tr')
rows

# Here we are Finding all the common attribute from the table that is table row(tr).
```

```python
currency_info=[]
for i in rows:
    currency=i.find_all('td')
#    print(currency)
    value_in_other_country=[c.text for c in currency]
#    print(value_in_other_country)
    currency_info.append(value_in_other_country)
    
# we are creating a empty list and initalizing for loop to get the data of the table. 
```

```python
currency_info
```

```python
currency_info=[]
for i in rows:
    currency=i.find_all('td')
    #print(currency)
    value_in_other_country=[c.text.replace('\n','') for c in currency]
    #print(value_in_other_country)
    currency_info.append(value_in_other_country[:-1])
    
# Initalizing for loop to get the data of the table and we are finding the all common attribute table data(td) from the table,
# Here we are replaceing \n by the replace method which is extra information in the table and then we append the data.
```

```python
currency_info
```

```python
Country=[]
Currency=[]
Spot=[]

for i in currency_info:
    Country.append(i[0])
    Currency.append(i[1])
    Spot.append(i[2])
    
# Assign Empty Lists with particular names and later on these lists will store the data. 
```

```python
currency_data={'Country_Currency_Code':Country, 'Currency_Name':Currency, 'Currencies_Value_by_Euro':Spot}
```

```python
converted_currency_data=pd.DataFrame(currency_data) 

# Here we are connecting the data into dataframe to get the data into table formate.
```
